### **Proses Analysis**

#### **Statistik Deskriptif**

##### **Correlation Analysis**

to **measure the strength of relationship between variables**, we are using the `kendall correlation` technique as we are dealing with an **ordinal variables**

the correlation coeficients range from -1 to 1 with variables values close to:
- `+1` indicate a **strong positive correlation**
- `-1` indicate a **strong negative correlation**
- `0` indicate a **weak correlation** 

and the p-value of the variables indicates:
- `pval>0.05` the correlation of between the two variables is **by chance**
- `pval<0.05` the correlation of between the two variables is **real**

```python
# Looping tiap kolom ordinal untuk menghitung positive correlation
for col in ordinal:
  if col != 'is_canceled':
   tmp = data.groupby(col)[['is_canceled']].mean().reset_index()
   tau, pval = stats.kendalltau(tmp[col],tmp['is_canceled'])
   # Menampilkan kolom yang memiliki korelasi tinggi dengan kolom is canceled
   if tau > 0.5:
      print('kendall correlation of',col,':',tau)
      print('p-value of',col,':',pval,"\n")
```

the kendall value show that **guest cancellation** has a `strong positive correlation` to:
- the **year** of the check in period
- the number of **adults** staying per room

but the p-value of the data shows that:
- the correlation between cancellation and year is **by chance**
- the correlation between cancellation and adults is **real**

```python
# Looping tiap kolom ordinal untuk menghitung negative correlation
for col in ordinal:
  if col != 'is_canceled':
   tmp = data.groupby(col)[['is_canceled']].mean().reset_index()
   tau, pval = stats.kendalltau(tmp[col],tmp['is_canceled'])
   # Menampilkan kolom yang memiliki korelasi tinggi dengan kolom is canceled
   if tau < -0.5:
      print('kendall correlation of',col,':',tau)
      print('p-value of',col,':',pval,"\n")
```

the kendall value show that **guest cancellation** has a `strong negative correlation` to:
- the number of **babies** staying per room
- the number of **repeated guest** staying per room

but the p-value of the data shows that:
- the correlation between cancellation and babies is **real**
- the correlation between cancellation and repeated guest is **by chance**

#### **Statistik Inferensial**

##### **Hypothesis Testing**

to **measure the relationship between variables**, we are using the `Chi-Squared` test as we are dealing with two **categorical variables**

Hypothesis:

`H0` : There is no relationship among variables

`H1` : There is relationship among variables

we will use confidence level of 95%, hence:
- if the `p-value<0.05`, the **H0** will be rejected
- if the `p-value>0.05`, the **H0** will be accepted

```python
# Looping tiap kolom kategorikal untuk menghitung p-value
for col in categorical:
  cross = pd.crosstab(data[col],data['is_canceled'].map({0:'No',1:'Yes'}))
  pval = stats.chi2_contingency(cross).pvalue
  # Print hasil hypothesis dari perhitungan chi-square test
  prompt = "-> H0 rejected" if pval <0.05 else "-> H0 accepted"
  if pval != 0.0:
    print(f"p-value of {col} and is_canceled: {pval} {prompt} \n")
```

`H0 rejected` means that is a **relationship between variables**

from this, we can see the factors that relate with guest cancellation is:
- Hotel types
- Arrival Month
- Hotel meal
- Reserved room type
- Assigned room type
- Customer type
